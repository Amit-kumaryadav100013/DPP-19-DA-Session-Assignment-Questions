**Question 5
Use Conditional Formatting to highlight rows where "Category" is "Electronics" and "Total Price" is greater than
$1000 ?. (Hint: Use formula in new rule)**


**Answer:-**   

          Steps To be Followed:
          Step 1: Select the entire data range (all rows and columns).
          Step 2: Go to Home then select the Conditional Formatting and select the New Rule.
          Step 3: Click “Use a formula to determine which cells to format”.
          Step 4: Enter this formula: =AND(C2="Electronics",E2>1000).
          Step 5: Click Format and Fill Color (e.g., Green) then Click OK.
